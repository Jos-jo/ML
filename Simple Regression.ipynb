```python
#importing the dataset
dataset = pd.read_csv('Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
#splitting the dataset into training and testing set  
#training 80% of the dataset and using testing 20%,hence 0.2 test_size.

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#training simple linear regression model

from sklearn.linear_model import LinearRegression 
# create a variable "regress", it creates a simple regression model

regress = LinearRegression()

#the fit method is used to train any regression model, method of the linear regression class

regress.fit(x_train, y_train)
```

```python
#Predicting testset result
y_predicted = regress.predict(x_test)
```

```python
#visualizing the trainimg test results with a 2D plot, plotting a graph btw real salaries and predicted salaries
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regress.predict(x_train), color = 'blue') #this plots the regression line
plt.title('Salary vs  Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()

```

```python
#visualizing the test results with a 2D plot, plotting a graph btw real salaries and predicted salaries

plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regress.predict(x_train), color = 'blue') #this plots the regression line and its  not changed here cos, its the same equation for the line.
plt.title('Salary vs  Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
